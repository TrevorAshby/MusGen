```python
cl_songs = pd.read_csv('classified_songs_V2.csv', delimiter='\t')
ID_TO_KEY = dict(zip(cl_songs.Id, cl_songs.Key))
DIAT_NOTES = {'C': 0, 'C#|Db': 1, 'D': 2, 'D#|Eb': 3, 'E': 4, 'F': 5, 
                    'F#|Gb': 6, 'G': 7, 'G#|Ab': 8, 'A': 9, 'A#|Bb': 10, 'B': 11}
NOTE_DISAMBIG = {'C': 'C', 'C#': 'C#|Db', 'Db': 'C#|Db', 'D': 'D', 'D#': 'D#|Eb', 'Eb': 'D#|Eb', 
                 'E':'E', 'F':'F', 'F#':'F#|Gb', 'Gb':'F#|Gb', 'G':'G', 'G#':'G#|Ab', 'Ab':'G#|Ab', 
                 'A':'A', 'A#':'A#|Bb', 'Bb':'A#|Bb', 'B':'B'}
key_to_flat_key = {'C': 'C', 'C#|Db': 'Db', 'Db': 'Db', 'D': 'D', 'D#|Eb': 'Eb', 'Eb': 'Eb', 
                 'E':'E', 'F':'F', 'F#|Gb':'Gb', 'Gb':'Gb', 'G':'G', 'G#|Ab':'Ab', 'Ab':'Ab', 
                 'A':'A', 'A#|Bb':'Bb', 'Bb':'Bb', 'B':'B'}

def note_to_int(note):
    return DIAT_NOTES[note]

def get_key(filename):
    return ID_TO_KEY[filename]

def get_tonic(key):
    # print('Implement func get_tonic')
    return NOTE_DISAMBIG[key.split('_')[0]]

def get_mode(key):
    # print('Implement func get_mode')
    return key.split('_')[-1]
```

```python
count = 0
vocab = MusicVocab.create()
src_dir_path = 'C:\\Users\\bkweb\\Documents\\dragn\\MusGen\\src\\numpy_path'
out_dir_path = 'C:\\Users\\bkweb\\Documents\\dragn\\MusGen\\src\\transposed'
not_found = []
for file in os.listdir(src_dir_path):
    arr = np.load(src_dir_path + '\\' + file)
    filename = file[:-4]
    print(file[:-4])
    mi = MusicItem.from_npenc(arr, vocab)
    if filename in ID_TO_KEY.keys():
        orig_key = get_key(filename)
        orig_tonic, orig_mode = get_tonic(orig_key), get_mode(orig_key)
        for i in range(12):
            new_tonic = int_to_note((note_to_int(orig_tonic) + i) % 12)
            print(f"\tnew_key: {new_tonic} {orig_mode}")
            out_path = out_dir_path + '\\' + filename + key_to_flat_key[new_tonic] + '_' + orig_mode + '.npy'
            np.save(out_path, mi.transpose(i).to_npenc())
    else:
        not_found.append(filename)
print("These files weren't found in ID_TO_KEY:")
print(not_found)
```
